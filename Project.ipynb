```python
# Importing the libraries
import cv2
import numpy as np
from PIL import Image, ImageTk
from skimage.util import random_noise 
from skimage.restoration import denoise_tv_chambolle, denoise_tv_bregman, denoise_bilateral
import tkinter as tk
from tkinter import *
from tkinter import filedialog
from tkinter.filedialog import askopenfile


"""
#############################
#  UI Design to take input  #
#############################
"""


# UI Design
input_win = tk.Tk()
input_win.geometry("800x400") 
input_win.title('Welcome')
input_font=('times', 36, 'bold')
l1 = tk.Label(input_win,text='Upload a color image',width=30,font=input_font)  
l1.grid(row=1,column=1,columnspan=4)
b1 = tk.Button(input_win, text='Upload File', width=26,command = lambda:upload_file())
b1.grid(row=2,column=1,columnspan=4)
b2 = tk.Button(input_win, text='Select a statistics-based filter', width=26, command = input_win.destroy)
b2.grid(row=3,column=1,columnspan=4)

def upload_file(): 
    global filename
    f_types = [('JPEG files', '*.jpg'), ('Portable Network Graphics','*.png'), 
               ('TIFF files', '*.tiff'), ('TIFF files', '*.tif')] # Needs update.
    filename = filedialog.askopenfilename(filetypes=f_types)
    img = ImageTk.PhotoImage(file=filename)
input_win.mainloop()

# Input image
img = cv2.imread(str(filename))


"""
#############################
#  Statistics-Based Filter  #
#############################
"""


# Take input
while True:
    try:
        act_1 = int(input('\nInput an integer to apply statistics-based filter:\n1. Arithmetic Mean\n2. Median\n3. Minimum\n4. Maximum\n5. Geometric Mean\n6. Harmonic Mean\n7. Contraharmonic Mean\n'))
    
    except ValueError:
        print('\nPlease, provide an integer value between 1 and 7.\n')
        continue
        
    if act_1>0 and act_1<8:
        print('\nInput successful!!!\n')
        break;
    
    else:
        print("\nPlease, input an integer value between 1 and 7.\n") 
        continue

        
if act_1 == 1:
    ############################
    #  Arithmetic Mean Filter  #
    ############################
    
    # Defining function to perform arithmetic mean filtering
    def Arithmetic_Mean_Filter(img, w):
        """Arithmetic Mean Filter Function"""
        for i in range(2,img.shape[0]-2):
            for j in range(2,img.shape[1]-2):
                blk = img[i-w:i+w+1, j-w:j+w+1]
                m = np.mean(blk,dtype=np.float32)
                img[i][j] = int(m)
        return img
    
    # Take amount from user
    while True:
        try:
            a = int(input('\nInput amount [0-2]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 0 and 2.\n')
            continue
        
        if a>-1 and a<3:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 0 and 2.\n") 
            continue

    # Spliting image
    b,g,r = cv2.split(img)

    # Calling function to perform arithmetic mean filtering
    b = Arithmetic_Mean_Filter(b, a)
    g = Arithmetic_Mean_Filter(g, a)
    r = Arithmetic_Mean_Filter(r, a)

    # Merging the filtered channels
    img_merge = cv2.merge([b, g, r])

    # Saving image
    cv2.imwrite('Arithmetic Mean Filter.png', img_merge)

    # Showing image
    cv2.imshow('Arithmetic Mean Filter', img_merge)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
elif act_1 == 2:
    ###################
    #  Median Filter  #
    ###################
    
    # Take amount from user
    while True:
        try:
            k = int(input('\nInput amount [odd integer between 1 and 715]:\n'))
    
        except ValueError:
            print('\nPlease, provide an odd integer value between 1 and 715.\n')
            continue
        
        if k>0 and k%2!=0 and k<717 :
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an odd integer value between 1 and 715.\n") 
            continue
            
    # Median Filter
    median_filter = cv2.medianBlur(img, k)

    # Saving image
    cv2.imwrite('Median Filter.png', median_filter)

    # Showing image
    cv2.imshow('Median Filter', median_filter)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
elif act_1 == 3:
    #####################
    #  Minimum Filter  #
    ####################

    # Defining function to perform minimum filtering
    def Minimum_Filter(image, n):
        """Minimum Filter Function"""
        # Create kernel
        size = (n, n)
        shape = cv2.MORPH_RECT
        kernel = cv2.getStructuringElement(shape, size)
        # Apply minimum filter with kernel
        output = cv2.erode(image, kernel)
        return output
    
    # Take amount from user
    while True:
        try:
            k = int(input('\nInput amount [odd integer between 1 and 999]:\n'))
    
        except ValueError:
            print('\nPlease, provide an odd integer value between 1 and 999.\n')
            continue
        
        if k>0 and k%2!=0 and k<1001 :
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an odd integer value between 1 and 999.\n") 
            continue

    # Calling function to perform minimum filtering
    minimum_filter = Minimum_Filter(img, k)

    # Saving image
    cv2.imwrite('Minimum Filter.png', minimum_filter)

    # Showing image
    cv2.imshow('Minimum Filter', minimum_filter)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
elif act_1 == 4:
    #####################
    #  Maximum Filter  #
    ####################
    
    # Defining function to perform maximum filtering
    def Maximum_Filter(image, n):
        """Maximum Filter Function"""
        # Create kernel
        size = (n,n)
        shape = cv2.MORPH_RECT
        kernel = cv2.getStructuringElement(shape, size)

        # Appliy maximum filter with kernel
        output = cv2.dilate(image, kernel)
        return output
    
    # Take amount from user
    while True:
        try:
            k = int(input('\nInput amount [1 - 1000]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 1 and 1000.\n')
            continue
        
        if k>0 and k<1001:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 1 and 1000.\n") 
            continue
        
    # Calling function to perform maximum filtering
    maximun_filter = Maximum_Filter(img, k)

    # Saving image
    cv2.imwrite('Maximum Filter.png', maximun_filter)

    # Showing image
    cv2.imshow('Maximum Filter', maximun_filter)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
elif act_1 == 5:
    ###########################
    #  Geometric Mean Filter  #
    ###########################
    
    # Defining function to geometric mean filter
    def Geometric_Mean_Filter(image, k):
        """Geometric Mean Filter Function"""
        rows, cols = image.shape[:2]
        padsize = int((k-1)/2)
        pad_img = cv2.copyMakeBorder(image, *[padsize]*4, cv2.BORDER_DEFAULT)
        geomean = np.zeros_like(image)
        for r in range(rows):
            for c in range(cols):
                geomean[r, c] = np.prod(pad_img[r:r+k, c:c+k])**(1/(k**2))
        geomean = np.uint8(geomean)
        norm = np.zeros((rows,cols))
        normalize = cv2.normalize(geomean,  norm, 0, 255, cv2.NORM_MINMAX)
        return normalize
    
    # Take amount from user
    while True:
        try:
            k = int(input('\nInput amount [1-3]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 1 and 3.\n')
            continue
        
        if k>0 and k<4 :
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 1 and 3.\n") 
            continue
    
    # Spliting image
    b,g,r = cv2.split(img)

    # Calling function to perform geometric mean filtering
    b = Geometric_Mean_Filter(b, k)
    g = Geometric_Mean_Filter(g, k)
    r = Geometric_Mean_Filter(r, k)

    # Merging the filtered channels
    geometric_mean = cv2.merge([b, g, r])

    # Saving image
    cv2.imwrite('Geometric Mean Filter.png', geometric_mean)

    # Showing image
    cv2.imshow('Geometric Mean Filter', geometric_mean)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
elif act_1 == 6:
    ##########################
    #  Harmonic Mean Filter  #
    ##########################
    
    # Defining function to harmonic mean filter
    def Harmonic_Mean_Filter(image, k):
        """Harmonic Mean Filter Function"""
        return round(len(image) / np.sum(1 / np.array(image)), k)

    # Take amount from user
    while True:
        try:
            k = int(input('\nInput amount [1-300]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 1 and 300.\n')
            continue
        
        if k>0 and k<301:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 1 and 300.\n") 
            continue
            
    # Spliting image
    b,g,r = cv2.split(img)

    # Calling function to perform harmonic mean filtering
    b = Harmonic_Mean_Filter(b, k)
    g = Harmonic_Mean_Filter(g, k)
    r = Harmonic_Mean_Filter(r, k)

    # Merging the filtered channels
    harmonic_mean = cv2.merge([b, g, r])

    # Saving image
    scale_abs = cv2.convertScaleAbs(harmonic_mean, alpha=(255.0))
    cv2.imwrite('Harmonic Mean Filter.png', scale_abs)

    # Showing image
    cv2.imshow('Harmonic Mean Filter', harmonic_mean)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
else:
    ################################
    #  Contraharmonic Mean Filter  #
    ################################
    
    # Defining function to contraharmonic mean filter
    def Contra_Harmonic_Mean(image):
        """Contraharmonic Mean Filter"""
        return round(np.sum(np.square(np.array(image))) / np.sum(np.array(image)))

    # Spliting image
    b,g,r = cv2.split(img)

    # Calling function to perform contraharmonic mean filtering
    b = Contra_Harmonic_Mean(b)
    g = Contra_Harmonic_Mean(g)
    r = Contra_Harmonic_Mean(r)

    # Merging the filtered channels
    contra_harmonic_mean = cv2.merge([b, g, r])

    # Saving image
    scale_abs = cv2.convertScaleAbs(contra_harmonic_mean, alpha=(255.0))
    cv2.imwrite('Contraharmonic Mean Filter.png', scale_abs)

    # Showing image
    cv2.imshow('Contraharmonic Mean Filter', contra_harmonic_mean)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
"""
#########################
#  Add simulated noise  #
#########################
"""


# Take input
while True:
    try:
        act_2 = int(input('\nInput an integer to add simulated noise:\n8. White noise\n9. Uniform noise\n10. “Salt and Pepper” noise\n11. Gaussian Noise\n12. Rayleigh Noise\n13. Erlang Noise\n14. Exponential Noise\n15. Sinusoidal Noise\n'))
    
    except ValueError:
        print('\nPlease, provide an integer value between 8 and 15.\n')
        continue
        
    if act_2>7 and act_2<16:
        print('\nInput successful!!!\n')
        break;
    
    else:
        print("\nPlease, input an integer value between 8 and 15.\n") 
        continue


if act_2 == 8:
    #################
    #  White Noise  #
    #################
    
    # Take magnitude from user
    while True:
        try:
            k = float(input('\nInput magnitude [0 - 1]:\n'))
    
        except ValueError:
            print('\nPlease, provide an float value between 0 and 1.\n')
            continue
        
        if k>=0 and k<=1:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an float value between 0 and 1.\n") 
            continue
            
    # White noise of a specified magnitude
    white_noised_image = random_noise(img, mode='salt', amount=k)

    # Saving image
    scale_abs = cv2.convertScaleAbs(white_noised_image, alpha=(255.0))
    cv2.imwrite('White Noised Image.png', scale_abs)

    # Showing image
    cv2.imshow('White Noised Image.png', white_noised_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


elif act_2 == 9:
    ###################
    #  Uniform Noise  #
    ###################

    # Defining function to add uniform noise
    def Uniform_Noise(image, a, b):
        """Uniform Noise Function"""
        x, y = image.shape
        n = np.zeros((x,y), dtype = np.float64)
        for i in range(x):
            for j in range(y):
                n[i][j] = np.random.uniform(a, b)
        noisey_img = image + n
        norm = np.zeros((x, y))
        normalize = cv2.normalize(noisey_img, norm, 0, 1, cv2.NORM_MINMAX)
        return normalize

    # Take magnitude from user
    while True:
        try:
            k = int(input('\nInput magnitude [0 - 1000]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 0 and 1000.\n')
            continue
        
        if k>=0 and k<1001:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 0 and 1000.\n") 
            continue
    
    # Take frequency from user
    while True:
        try:
            l = int(input('\nInput frequency [0 - 50]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 0 and 50.\n')
            continue
        
        if l>=0 and l<51:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 0 and 50.\n") 
            continue
            
    # Spliting image to add uniform noise
    b,g,r = cv2.split(img)

    # Calling function to add uniform noise
    b = Uniform_Noise(b, k, l)
    g = Uniform_Noise(g, k, l)
    r = Uniform_Noise(r, k, l)

    # Merging the filtered channels of the image
    uniform_noise = cv2.merge([b, g, r])

    # Saving image
    scale_abs = cv2.convertScaleAbs(uniform_noise, alpha=(255.0))
    cv2.imwrite('Uniform Noised Image.png', scale_abs)

    # Showing image
    cv2.imshow('Uniform Noised Image.png', uniform_noise)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


elif act_2 == 10:
    #############################
    #  “Salt and Pepper” Noise  #
    #############################
    
    # Take magnitude from user
    while True:
        try:
            k = float(input('\nInput magnitude [0 - 1]:\n'))
    
        except ValueError:
            print('\nPlease, provide an float value between 0 and 1.\n')
            continue
        
        if k>=0 and k<=1:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an float value between 0 and 1.\n") 
            continue
    
    # “Salt and Pepper” noise of a specified magnitude and probability
    s_p_noised_image = random_noise(img, mode='s&p', amount=k)

    # Saving image
    scale_abs = cv2.convertScaleAbs(s_p_noised_image, alpha=(255.0))
    cv2.imwrite('Salt and Pepper Noised Image.png', scale_abs)

    # Showing image
    cv2.imshow('Salt and Pepper Noised Image.png', s_p_noised_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
elif act_2 == 11:
    ####################
    #  Gaussian Noise  #
    ####################
    
    # Gaussian Noise
    gaussian_noise = random_noise(img, mode='gaussian')

    # Saving image
    scale_abs = cv2.convertScaleAbs(gaussian_noise, alpha=(255.0))
    cv2.imwrite('Gaussian Noised Image.png', scale_abs)

    # Showing image
    cv2.imshow('Gaussian Noised Image.png', gaussian_noise)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


elif act_2 == 12:
    ####################
    #  Rayleigh Noise  #
    ####################
    
    # Take amount from user
    while True:
        try:
            k = float(input('\nInput amount [0 - 2]:\n'))
    
        except ValueError:
            print('\nPlease, provide an float value between 0 and 2.\n')
            continue
        
        if k>=0 and k<=2:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an float value between 0 and 2.\n") 
            continue
            
    # Rayleigh Noise
    image_r = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    image_r = image_r.astype(np.float64)
    noise_std = k
    noise = np.random.rayleigh(noise_std, image_r.shape)
    rayleigh_noiseed_image = cv2.addWeighted(image_r, 1, noise, 70, 0.0).astype(np.uint8)

    # Saving image
    scale_abs = rayleigh_noiseed_image
    cv2.imwrite('Rayleigh Noised Image.png', scale_abs)

    # Showing image
    cv2.imshow('Rayleigh Noised Image.png', rayleigh_noiseed_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


elif act_2 == 13:
    ##################
    #  Erlang Noise  #
    ##################

    # Defining function to add erlang noise
    def Erlang_Noise_Function(image, shape, scale):
        """Erlang Noise Function"""
        noise = np.random.gamma(shape, scale, image.shape)
        noisy_image = cv2.add(image, np.uint8(noise))
        return noisy_image

    # Take magnitude from user
    while True:
        try:
            k = int(input('\nInput magnitude [0 - 2000]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 0 and 2000.\n')
            continue
        
        if k>=0 and k<2001:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 0 and 2000.\n") 
            continue
    
    # Take frequency from user
    while True:
        try:
            l = int(input('\nInput frequency [0 - 100]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 0 and 50.\n')
            continue
        
        if l>=0 and l<101:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 0 and 50.\n") 
            continue
            
    # Spliting image to add erlang noise
    b,g,r = cv2.split(img)

    # Calling function to add erlang noise
    b = Erlang_Noise_Function(b, k, l)
    g = Erlang_Noise_Function(g, k, l)
    r = Erlang_Noise_Function(r, k, l)

    # Merging the filtered channels of the image
    erlang_noised_image = cv2.merge([b, g, r])

    # Saving image
    scale_abs = erlang_noised_image
    cv2.imwrite('Erlang Noised Image.png', erlang_noised_image)

    # Showing image
    cv2.imshow('Erlang Noised Image.png', erlang_noised_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
elif act_2 == 14:
    #######################
    #  Exponential Noise  #
    #######################
    
    # Defining function to add exponential noise
    def Exponential_Noise_Function(image, scale):
        """Exponential Noise Function"""
        size = image.shape
        noise = np.random.exponential(scale, size)
        noisy_image = cv2.add(image, np.uint8(noise))
        return noisy_image

    # Take amount from user
    while True:
        try:
            k = int(input('\nInput amount [0 - 2000]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 0 and 2000.\n')
            continue
        
        if k>=0 and k<2001:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 0 and 2000.\n") 
            continue
            
    # Spliting image
    b,g,r = cv2.split(img)

    # Calling function to add exponential noise
    b = Exponential_Noise_Function(b, k)
    g = Exponential_Noise_Function(g, k)
    r = Exponential_Noise_Function(r, k)

    # Merging the filtered channels of the image
    exponential_noised_image = cv2.merge([b, g, r])

    # Saving image
    scale_abs = exponential_noised_image
    cv2.imwrite('Exponential Noised Image.png', exponential_noised_image)

    # Showing image
    cv2.imshow('Exponential Noised Image.png', exponential_noised_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
else :
    ######################
    #  Sinusoidal Noise  #
    ######################
    
    # Defining function to perform sinusoidal noise
    def Sinusoidal_Noise_Function(image, amp, freq):
        """Sinusoidal Noise Function"""
        row, col = image.shape
        x = np.linspace(0, 2*np.pi, row)
        y = np.linspace(0, 2*np.pi, col)
        x,y = np.meshgrid(x,y)
        sin_wave = amp*np.sin(freq*x + freq*y)
        output = image + sin_wave.astype(np.uint8)
        return output
    
    # Take amplitude from user
    while True:
        try:
            k = int(input('\nInput amplitude [0 - 50]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 0 and 50.\n')
            continue
        
        if k>=0 and k<51:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 0 and 50.\n") 
            continue
            
    # Take frequency from user
    while True:
        try:
            l = int(input('\nInput frequency [0 - 100]:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 0 and 100.\n')
            continue
        
        if l>=0 and l<101:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 0 and 100.\n") 
            continue
            
    # Spliting image
    b,g,r = cv2.split(img)

    # Calling function to add exponential noise
    b = Sinusoidal_Noise_Function(b, k, l)
    g = Sinusoidal_Noise_Function(g, k, l)
    r = Sinusoidal_Noise_Function(r, k, l)

    # Merging the filtered channels of the image
    sinusoidal_noised_image = cv2.merge([b, g, r])

    # Saving image
    scale_abs = sinusoidal_noised_image
    cv2.imwrite('Sinusoidal Noised Image.png', sinusoidal_noised_image)

    # Showing image
    cv2.imshow('Sinusoidal Noised Image.png', sinusoidal_noised_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


"""
##################
#  Remove Noise  #
##################
"""


# Take input
while True:
    try:
        act_3 = int(input('\nInput an integer to remove noise:\n16. White noise\n17. Uniform noise\n18. Gaussian noise\n19. Rayleigh noise\n20. Erlang noise\n21. Exponential noise\n'))
    
    except ValueError:
        print('\nPlease, provide an integer value between 16 and 21.\n')
        continue
        
    if act_3>15 and act_3<22:
        print('\nInput successful!!!\n')
        break;
    
    else:
        print("\nPlease, input an integer value between 16 and 21.\n") 
        continue
        
        
if act_3 == 16:
    ########################
    #  Remove White Noise  #
    ########################

    # Take amount from user
    while True:
        try:
            k = int(input('\nInput amount [odd integer between 1 and 13]:\n'))
    
        except ValueError:
            print('\nPlease, provide an odd integer value between 1 and 13.\n')
            continue
        
        if k>0 and k%2!=0 and k<15 :
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an odd integer value between 1 and 13.\n") 
            continue
            
    # Remove white noise of a specified magnitude
    remove_white_noise = cv2.medianBlur(scale_abs, k)

    # Saving image
    cv2.imwrite('White Noise Removed Image.png', remove_white_noise)

    # Showing image
    cv2.imshow('White Noise Removed Image.png', remove_white_noise)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


elif act_3 == 17:
    ##########################
    #  Remove Uniform Noise  #
    ##########################
    
    # Take amount from user
    while True:
        try:
            k = int(input('\nInput amount [odd integer between 1 and 19]:\n'))
    
        except ValueError:
            print('\nPlease, provide an odd integer value between 1 and 19.\n')
            continue
        
        if k>0 and k%2!=0 and k<21 :
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an odd integer value between 1 and 19.\n") 
            continue
            
    # Remove uniform noise of a specified magnitude and bandwidth
    remove_uniform_noise = cv2.GaussianBlur(scale_abs,(k,k),0)

    # Saving image
    cv2.imwrite('Uniform Noise Removed Image.png', remove_uniform_noise)

    # Showing image
    cv2.imshow('Uniform Noise Removed Image.png', remove_uniform_noise)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


elif act_3 == 18:
    ###########################
    #  Remove Gaussian Noise  #
    ###########################
    
    # Remove gaussian noise
    remove_gaussian_noise = cv2.fastNlMeansDenoisingColored(scale_abs,None,10,10,9,33)

    # Saving image
    cv2.imwrite('Gaussian Noise Removed Image.png', remove_gaussian_noise)

    # Showing image
    cv2.imshow('Gaussian Noise Removed Image.png', remove_gaussian_noise)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


elif act_3 == 19:
    ###########################
    #  Remove Rayleigh Noise  #
    ###########################
    
    # Take amount from user
    while True:
        try:
            k = float(input('\nInput amount [0.1 - 2]:\n'))
    
        except ValueError:
            print('\nPlease, provide an float value between 0.1 and 2.\n')
            continue
        
        if k>0 and k<=2:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an float value between 0.1 and 2.\n") 
            continue
            
    # Remove rayleigh noise
    remove_rayleigh_noise = denoise_tv_chambolle(scale_abs, weight=k, channel_axis=-1, eps=0.0002, max_num_iter=200)

    # Saving image
    scale_abs = cv2.convertScaleAbs(remove_rayleigh_noise, alpha=(255.0))
    cv2.imwrite('Rayleigh Noise Removed Image.png', scale_abs)

    # Showing image
    cv2.imshow('Rayleigh Noise Removed Image.png', remove_rayleigh_noise)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


elif act_3 == 20:
    #########################
    #  Remove Erlang Noise  #
    #########################
    
    # Take amount from user
    while True:
        try:
            k = float(input('\nInput amount [0.1 - 9]:\n'))
    
        except ValueError:
            print('\nPlease, provide an float value between 0.1 and 9.\n')
            continue
        
        if k>0 and k<=9:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an float value between 0.1 and 9.\n") 
            continue
            
    # Remove erlang noise
    remove_erlang_noise = denoise_tv_bregman(scale_abs, weight=k, max_num_iter=100, eps=0.001, isotropic=True, channel_axis=-1)

    # Saving image
    scale_abs = cv2.convertScaleAbs(remove_erlang_noise, alpha=(255.0))
    cv2.imwrite('Erlang Noise Removed Image.png', scale_abs)

    # Showing image
    cv2.imshow('Erlang Noise Removed Image.png', remove_erlang_noise)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
else :
    ##############################
    #  Remove Exponential Noise  #
    ##############################
    
    # Take amount from user
    while True:
        try:
            k = int(input('\nInput amount integer between 0 and 25:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between 0 and 25.\n')
            continue
        
        if k>=0 and k<26 :
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between 1 and 25.\n") 
            continue
    
    # Remove exponential noise
    remove_exponential_noise = denoise_bilateral(scale_abs, win_size=k, sigma_color=None, sigma_spatial=1, bins=10000, mode='constant', cval=0, channel_axis=-1)

    # Saving image
    scale_abs = cv2.convertScaleAbs(remove_exponential_noise, alpha=(255.0))
    cv2.imwrite('Exponential Noise Removed Image.png', scale_abs)

    # Showing image
    cv2.imshow('Exponential Noise Removed Image.png', remove_exponential_noise)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
"""
######################
#  Color Adjustment  #
######################
"""


# Take input
while True:
    try:
        act_4 = int(input('\nHow to do color adjustment:\n22. Color negative\n23. Display of Red, Green, and Blue planes separately\n24. Display of Cyan, Magenta, and Yellow planes separately\n25. Display of Y, Cb, and Cr planes separately\n26. Display of H, S, and V planes separately\n27. Individual adjustments of R, G, and B components\n'))
    
    except ValueError:
        print('\nPlease, provide an integer value between 22 and 27.\n')
        continue
        
    if act_4>21 and act_4<28:
        print('\nInput successful!!!\n')
        break;
    
    else:
        print("\nPlease, input an integer value between 22 and 27.\n") 
        continue    

        
if act_4 == 22:
    ####################
    #  Color Negative  #
    ####################
    
    # Color negative (also converts RGB to CMY)
    # Covert image to RGB
    image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Spliting image
    r,g,b = cv2.split(image)
    
    # C, M, Y conversion
    c = 1 - r/255
    m = 1 - g/255
    y = 1 - b/255

    # Merging the channels of the image
    img_merge = cv2.merge([c, m, y])
    img_merge = cv2.convertScaleAbs(img_merge, alpha=(255.0))
    
    # Saving image
    cv2.imwrite('Color Negative.png', img_merge)

    # Showing image
    cv2.imshow('Color Negative', img_merge)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
elif act_4 == 23:
    #######################################################
    #  Display of Red, Green, and Blue planes separately  #
    #######################################################
    
    # Spliting image
    b,g,r = cv2.split(img)
    
    # define channel zeros channel
    zeros = np.zeros(b.shape, np.uint8)
    
    # Make BGR image
    b_BGR = cv2.merge([b,zeros,zeros])
    g_BGR = cv2.merge([zeros,g,zeros])
    r_BGR = cv2.merge([zeros,zeros,r])
    
    # Saving image
    cv2.imwrite('Blue Channel.png', b_BGR)
    cv2.imwrite('Green Channel.png', g_BGR)
    cv2.imwrite('Red Channel.png', r_BGR)
    
    # display of Red, Green, and Blue planes separately
    cv2.imshow('Blue Channel', b_BGR)
    cv2.waitKey(0)
    cv2.imshow('Green Channel', g_BGR)
    cv2.waitKey(0)
    cv2.imshow('Red Channel', r_BGR)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    
elif act_4 == 24:
    ############################################################
    #  Display of Cyan, Magenta, and Yellow planes separately  #
    ############################################################
    
    # Covert image to RGB
    image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Spliting image
    r,g,b = cv2.split(image)
    
    # C, M, Y conversion
    c = 1 - r/255
    m = 1 - g/255
    y = 1 - b/255
    
    # Normalize the channels
    c = cv2.convertScaleAbs(c, alpha=(255.0))
    m = cv2.convertScaleAbs(m, alpha=(255.0))
    y = cv2.convertScaleAbs(y, alpha=(255.0))
    
    # Saving image
    cv2.imwrite('Cyan Channel.png', c)
    cv2.imwrite('Magenta Channel.png', m)
    cv2.imwrite('Yellow Channel.png', y)
    
    # display of Cyan, Magenta, and Yellow planes separately
    cv2.imshow('Cyan Channel', c)
    cv2.waitKey(0)
    cv2.imshow('Magenta Channel', m)
    cv2.waitKey(0)
    cv2.imshow('Yellow Channel', y)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


elif act_4 == 25:
    ################################################
    #  Display of Y, Cb, and Cr planes separately  #
    ################################################
    
    # Covert image to YCbCr
    image = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)

    # Spliting image
    Y,Cr,Cb = cv2.split(image)
    
    # Saving image
    cv2.imwrite('Y Channel.png', Y)
    cv2.imwrite('Cb Channel.png', Cb)
    cv2.imwrite('Cr Channel.png', Cr)
    
    # display of Cyan, Magenta, and Yellow planes separately
    cv2.imshow('Y Channel', Y)
    cv2.waitKey(0)
    cv2.imshow('Cb Channel', Cb)
    cv2.waitKey(0)
    cv2.imshow('Cr Channel', Cr)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    
elif act_4 == 26:
    ##############################################
    #  Display of H, S, and V planes separately  #
    ##############################################
    
    # Covert image to HSV
    image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Spliting image
    H,S,V = cv2.split(image)
    
    # Saving image
    cv2.imwrite('H Channel.png', H)
    cv2.imwrite('S Channel.png', S)
    cv2.imwrite('V Channel.png', V)
    
    # display of Cyan, Magenta, and Yellow planes separately
    cv2.imshow('H Channel', H)
    cv2.waitKey(0)
    cv2.imshow('S Channel', S)
    cv2.waitKey(0)
    cv2.imshow('V Channel', V)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    
else :
    ######################################################
    #  Individual adjustments of R, G, and B components  #
    ######################################################

    # Split image
    b,g,r = cv2.split(img)
    
    #Take R value from user
    while True:
        try:
            r_v = int(input('\nInput R value to adjust:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between -255 and 255.\n')
            continue
        
        if r_v>-256 and r_v<256:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between -255 and 255.\n") 
            continue 
    
    # Adjust R value
    if r_v>-256 and r_v<0:
        r = cv2.subtract(r, r_v)
    else:
        r = cv2.add(r, r_v)
        
        
    # Take G value from user
    while True:
        try:
            g_v = int(input('\nInput G value to adjust:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between -255 and 255.\n')
            continue
        
        if g_v>-256 and g_v<256:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between -255 and 255.\n") 
            continue 
    
    # Adjust G value
    if g_v>-256 and g_v<0:
        g = cv2.subtract(g, g_v)
    else:
        g = cv2.add(g, g_v)
        

    # Take B value from user
    while True:
        try:
            b_v = int(input('\nInput B value to adjust:\n'))
    
        except ValueError:
            print('\nPlease, provide an integer value between -255 and 255.\n')
            continue
        
        if b_v>-256 and b_v<256:
            print('\nInput successful!!!\n')
            break;
    
        else:
            print("\nPlease, input an integer value between -255 and 255.\n") 
            continue 
    
    # Adjust B value
    if b_v>-256 and b_v<0:
        b = cv2.subtract(b, b_v)
    else:
        b = cv2.add(b, b_v)
        
        
    # Merge channels
    img_merge = cv2.merge((b,g,r))

    # Saving image
    normalize = cv2.normalize(img_merge, None, 0, 255, cv2.NORM_MINMAX)
    cv2.imwrite('Adjusted image.png', normalize)
    
    # Showing image
    cv2.imshow('Adjusted image.png', normalize)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
